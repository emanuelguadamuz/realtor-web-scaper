##### Paso 7. Agregar Datos de Página a una Lista.

```python
id_propiedad = []
tipo_propiedad = []
direccion = []
habitaciones = [] 
baños = []
plazas_aparcamiento = []
tamaño_terreno = []
precio_lista = []
precio_consumidor = []
precio_seguimiento = []
nombre_agencia = []
telefono_oficina = []
sitio_web = []
nombre_agente = []
correo_electronico = []

for result in result_items:
    # ID de la Propiedad
    id_propiedad.append(result['id'])
    
    # Tipo de Propiedad
    tipo_propiedad.append(result['propertyTypes'])

    # Dirección
    direccion.append(result['displayAddress'])

    # Habitaciones
    habitaciones.append(result['bedrooms'])

    # Baños
    baños.append(result['bathrooms'])

    # Plazas de Aparcamiento
    plazas_aparcamiento.append(result['parkingSpaces'])

    # Tamaño del Terreno
    tamaño_terreno.append(result['landSize'] if result['landSize'] else None)

    # Precio de Lista
    precio_lista.append(result['price']['displayListingPrice'])
    
    # Precio al Consumidor
    precio_consumidor.append(result['price']['displayConsumerPrice'])

    # Precio de Seguimiento
    precio_seguimiento.append(result['price']['displayTrackingPrice'])

    # Nombre de la Agencia
    nombre_agencia.append(result['agents'][0]['agencyName'] if result['agents'] else None)

    # Telefono de Oficina
    telefono_oficina.append(result['agents'][0]['officePhone'] if result['agents'] else None)

    # Sitio Web
    sitio_web.append(result['agents'][0]['contact']['website'] if result['agents'] else None)
    
    # Nombre del Agente
    nombre_agente.append(result['agents'][0]['name'] if result['agents'] else None)

    # Correo Electrónico de Agente
    correo_electronico.append(result['agents'][0]['contact']['emails'][0] if result['agents'] and result['agents'][0]['contact']['emails'] else None)
```

##### Paso 8. Pandas: Marco de Datos

```python
bqt_properties_df = pd.DataFrame({
    'ID Propiedad': id_propiedad,
    'Tipo Propiedad': tipo_propiedad,
    'Dirección': direccion,
    'Habitaciones': habitaciones,
    'Baños': baños,
    'Plazas de Aparcamiento': plazas_aparcamiento,
    'Tamaño Terreno': tamaño_terreno,
    'Precio Lista': precio_lista,
    'Precio Consumidor': precio_consumidor,
    'Precio Seguimiento': precio_seguimiento,
    'Nombre Agencia': nombre_agencia,
    'Teléfono Oficina': telefono_oficina,
    'Sitio Web': sitio_web,
    'Nombre Agente': nombre_agente,
    'Correo Electrónico Agente': correo_electronico
})
```
